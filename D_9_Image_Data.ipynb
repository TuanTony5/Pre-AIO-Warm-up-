```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Increase brightness
img = cv2.imread("image.jpg", 1)

img = img.astype(float)
img += 50

img = np.clip(img, a_min=0, a_max=255)
img = img.astype(np.uint8)

plt.imshow(img)
```

```python
# Decrease brightness
img = cv2.imread("image.jpg", 1)

img = img.astype(float)
img += -80

img = np.clip(img, a_min=0, a_max=255)
img = img.astype(np.uint8)

plt.imshow(img)
```

```python
img = cv2.imread('image.jpg',0)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img[:,:115] = 255
plt.imshow(img)
```

```python
# Cropping a region
img = cv2.imread('image.jpg',1)
high,width,channel = img.shape

x0,y0 = 150,350
x1,y1 = 210,400
reg = img[y0:y1,x0:x1,:]
plt.imshow(reg)
```

```python
# Draw bounding box
color = [255,0,0]
img[y0:y1,x0] = color
img[y0:y1,x1] = color
img[y0,x0:x1] = color
# img[y1,x0:x1] = color

plt.imshow(img)
```

```python
# Convert color image to gray image
img = cv2.imread("image.jpg", 1) # B G R

vector = np.array([0.072, 0.715, 0.213])

gray_img = img.dot(vector)
gray_img = np.uint8(gray_img)
gray_img = cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB)
plt.imshow(gray_img)
```

```python
# Flip the image vertically
img = cv2.imread("image.jpg", 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
high, width, channel = img.shape
transform = np.array([[-1, 0], [0, 1]])

output_verti = np.zeros(img.shape).astype(np.uint8)

for x in range(high):
    for y in range(width):
        pixel = img[x,y,:]
        location = np.array([x,y])
        new_x,new_y = transform.dot(location) + [high-1,0]
        output_verti[new_x,new_y,:] = pixel
    
plt.imshow(output_verti)
```

```python
# Flip the image horizontally
transform = np.array([[1, 0], [0, -1]])
output_hori = np.zeros(img.shape).astype(np.uint8)
for x in range(high):
    for y in range(width):
        pixel = img[x,y,:]
        location = np.array([x,y])
        new_x,new_y = transform.dot(location) + [0,width-1]
        output_hori[new_x,new_y,:] = pixel
    
plt.imshow(output_hori)
```
