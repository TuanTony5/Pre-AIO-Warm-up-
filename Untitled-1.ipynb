```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

ob_img = cv2.imread("Background_Subtraction/vovinam.jpg", 0)
# ob_img = cv2.cvtColor(ob_img, cv2.COLOR_BGR2RGB)
plt.imshow(ob_img)

height,width = ob_img.shape
# create a 2D list
copy_image = [[0]*width  
              for _ in range(height)]
# copy
for i in range(height):
    for j in range(width):
        copy_image[i][j] = ob_img[i][j]-25


# save an image from a list
cv2.imwrite('copy.jpg', 
            np.array(copy_image))
```
