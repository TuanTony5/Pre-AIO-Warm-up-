# Electricity Transformer Dataset

```python
df_ETTh = pd.read_csv('https://raw.githubusercontent.com/zhouhaoyi/ETDataset/main/ETT-small/ETTh1.csv',parse_dates=['date'])
df_ETTh.head()
```

# Word Cloud

```python
from wordcloud import WordCloud,STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_fs = pd.read_csv("five_star_reviews.csv")
df_fs.head()

stopwords = set(STOPWORDS)
words = ""
for review in df_fs.Review:
    tokens = str(review).split()
    tokens = [i.lower() for i in tokens]
    words += " ".join(tokens) + " "

wordcloud = WordCloud(
    width=800,
    height=800,
    background_color="white",
    stopwords=stopwords,
    min_font_size=10,
).generate(words)

# Plot wordcloud image
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.title("Common words for 5 stars reviews")

plt.show()
```
