# Time-series data

```python
import pandas as pd 

df = pd.read_csv("Pandas_data/weatherHistory_univariate.csv")
df
```

```python
# Sort data by date columns
df = df.sort_values('Formatted Date')
df.head()
```

```python
# Convert to datetime object
df['Formatted Date'] = [pd.to_datetime(time) for time in df['Formatted Date']]
df.head()
```

```python
df['date'] = [time.date() for time in df['Formatted Date']]
df.head()
```

```python
df['day'] = [time.day for time in df['Formatted Date']]
df.head()
```

```python
df['timestamp'] = [time.timestamp() for time in df['Formatted Date']]
df.head()
```
