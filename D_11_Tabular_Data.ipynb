```python
import pandas as pd

# Creating data
data = {
    'TV': [100, 200, 300, 400, 500],
    'Radio': [50, 100, 150, 200, 250],
    'Phone': [20, 40, 60, 80, 100],
    'Sales': [200, 400, 600, 800, 1000]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Displaying the DataFrame
df
```

```python
import seaborn as sns
sns.jointplot(x='TV',y='Sales',data=df)
```
